# Cost Structure and Channel-Wise Profitability Analysis for Multi-Channel Restaurants

```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
pd.set_option("display.max_columns", None)
```

```python
# loading dataset
df = pd.read_csv("SkyCity Auckland Restaurants & Bars.csv")

print("Shape:", df.shape)
df.head()
```

```python
#data understanding 
df.info()

df.describe().T

# Missing values
df.isna().sum().sort_values(ascending=False)
```

```python
# data validation
# Project requirement: validate percentages and shares.

# Checking rate columns
rate_cols = ["COGSRate", "OPEXRate", "CommissionRate"]

for col in rate_cols:
    print(f"{col} range:", df[col].min(), "to", df[col].max())

# Validate channel share sums
df["ShareSum"] = (
    df["InStoreShare"]
    + df["UE_share"]
    + df["DD_share"]
    + df["SD_share"]
)

df["ShareSum"].describe()
```

```python
# Removing impossible revenue rows
df = df[df["MonthlyOrders"] > 0]

# Clipping extreme rates if present
for col in rate_cols:
    df[col] = df[col].clip(lower=0, upper=1)
```

### PROFIT RE-CALCULATION

```python
# ---------- COGS ----------
df["COGS_InStore"] = df["InStoreRevenue"] * df["COGSRate"]
df["COGS_UE"] = df["UberEatsRevenue"] * df["COGSRate"]
df["COGS_DD"] = df["DoorDashRevenue"] * df["COGSRate"]
df["COGS_SD"] = df["SelfDeliveryRevenue"] * df["COGSRate"]

# ---------- OPEX ----------
df["OPEX_InStore"] = df["InStoreRevenue"] * df["OPEXRate"]
df["OPEX_UE"] = df["UberEatsRevenue"] * df["OPEXRate"]
df["OPEX_DD"] = df["DoorDashRevenue"] * df["OPEXRate"]
df["OPEX_SD"] = df["SelfDeliveryRevenue"] * df["OPEXRate"]

# ---------- Commission ----------
df["Commission_UE"] = df["UberEatsRevenue"] * df["CommissionRate"]
df["Commission_DD"] = df["DoorDashRevenue"] * df["CommissionRate"]
```

```python
# recalculating net profit
# In-store
df["Calc_InStoreProfit"] = (
    df["InStoreRevenue"]
    - df["COGS_InStore"]
    - df["OPEX_InStore"]
)

# Uber Eats
df["Calc_UE_Profit"] = (
    df["UberEatsRevenue"]
    - df["COGS_UE"]
    - df["OPEX_UE"]
    - df["Commission_UE"]
)

# DoorDash
df["Calc_DD_Profit"] = (
    df["DoorDashRevenue"]
    - df["COGS_DD"]
    - df["OPEX_DD"]
    - df["Commission_DD"]
)

# Self Delivery
df["Calc_SD_Profit"] = (
    df["SelfDeliveryRevenue"]
    - df["COGS_SD"]
    - df["OPEX_SD"]
    - df["SD_DeliveryTotalCost"]
)
```

```python
# Profit Validation Check
profit_check = pd.DataFrame({
    "InStore_diff": df["InStoreNetProfit"] - df["Calc_InStoreProfit"],
    "UE_diff": df["UberEatsNetProfit"] - df["Calc_UE_Profit"],
    "DD_diff": df["DoorDashNetProfit"] - df["Calc_DD_Profit"],
    "SD_diff": df["SelfDeliveryNetProfit"] - df["Calc_SD_Profit"],
})

profit_check.describe()
```

### Channel Margin Analysis

```python
#margin%
df["Margin_InStore"] = df["Calc_InStoreProfit"] / df["InStoreRevenue"]
df["Margin_UE"] = df["Calc_UE_Profit"] / df["UberEatsRevenue"]
df["Margin_DD"] = df["Calc_DD_Profit"] / df["DoorDashRevenue"]
df["Margin_SD"] = df["Calc_SD_Profit"] / df["SelfDeliveryRevenue"]
```

```python
#visualization - Profit by channel

channel_profit = {
    "InStore": df["Calc_InStoreProfit"].sum(),
    "UberEats": df["Calc_UE_Profit"].sum(),
    "DoorDash": df["Calc_DD_Profit"].sum(),
    "SelfDelivery": df["Calc_SD_Profit"].sum(),
}

plt.figure(figsize=(8,5))
plt.bar(channel_profit.keys(), channel_profit.values())
plt.title("Total Profit by Channel")
plt.ylabel("Profit")
plt.show()
```

```python
# Profit per Order

# helper to find correct column
def find_col(possible_names):
    for name in possible_names:
        if name in df.columns:
            return name
    return None

ue_orders_col = find_col([
    "UberEatsOrdersCount",
    "UberEatsOrders"
])

dd_orders_col = find_col([
    "DoorDashOrdersCount",
    "DoorDashOrders"
])

sd_orders_col = find_col([
    "SelfDeliveryOrdersCount",
    "SelfDeliveryOrders"
])

print("UE orders column:", ue_orders_col)
print("DD orders column:", dd_orders_col)
print("SD orders column:", sd_orders_col)

# Profit per order
df["ProfitPerOrder_UE"] = df["Calc_UE_Profit"] / df[ue_orders_col]
df["ProfitPerOrder_DD"] = df["Calc_DD_Profit"] / df[dd_orders_col]
df["ProfitPerOrder_SD"] = df["Calc_SD_Profit"] / df[sd_orders_col]
```

```python
# Commission Sensitivity

# What-if commission increases
test_rates = [0.15, 0.20, 0.25, 0.30]

sensitivity = []

for r in test_rates:
    temp_profit = (
        df["UberEatsRevenue"]
        - df["COGS_UE"]
        - df["OPEX_UE"]
        - df["UberEatsRevenue"] * r
    ).sum()
    
    sensitivity.append(temp_profit)

pd.DataFrame({
    "CommissionRate": test_rates,
    "Total_UE_Profit": sensitivity
})
```

```python
# Cuisine Heatmap

pivot_margin = df.pivot_table(
    values="Margin_UE",
    index="CuisineType",
    columns="Segment",
    aggfunc="mean"
)

plt.figure(figsize=(10,6))
sns.heatmap(pivot_margin, annot=True, fmt=".2f", cmap="RdYlGn")
plt.title("Uber Eats Margin Heatmap")
plt.show()
```

```python
# Profit Volatility

margin_cols = ["Margin_InStore", "Margin_UE", "Margin_DD", "Margin_SD"]

df["ProfitVolatility"] = df[margin_cols].std(axis=1)

df["RiskFlag"] = pd.cut(
    df["ProfitVolatility"],
    bins=[-np.inf, 0.05, 0.15, np.inf],
    labels=["Stable", "Moderate Risk", "High Risk"]
)

df["RiskFlag"].value_counts()
```

# Advanced Profitability Analysis

```python
#Channel Efficiency Comparison

channel_summary = pd.DataFrame({
    "Total_Revenue": [
        df["InStoreRevenue"].sum(),
        df["UberEatsRevenue"].sum(),
        df["DoorDashRevenue"].sum(),
        df["SelfDeliveryRevenue"].sum(),
    ],
    "Total_Profit": [
        df["Calc_InStoreProfit"].sum(),
        df["Calc_UE_Profit"].sum(),
        df["Calc_DD_Profit"].sum(),
        df["Calc_SD_Profit"].sum(),
    ],
}, index=["InStore", "UberEats", "DoorDash", "SelfDelivery"])

channel_summary["Margin_%"] = (
    channel_summary["Total_Profit"] / channel_summary["Total_Revenue"]
)

channel_summary
```

```python
# Visualization — Margin Comparison

plt.figure(figsize=(8,5))
plt.bar(channel_summary.index, channel_summary["Margin_%"])
plt.title("Channel Margin Comparison")
plt.ylabel("Margin %")
plt.show()
```

```python
# Commission Drag Analysis

commission_impact = pd.DataFrame({
    "Channel": ["UberEats", "DoorDash"],
    "Total_Commission": [
        df["Commission_UE"].sum(),
        df["Commission_DD"].sum(),
    ]
})

commission_impact
```

```python
#commission vs profit chart

commission_vs_profit = pd.DataFrame({
    "UE": [
        df["Calc_UE_Profit"].sum(),
        df["Commission_UE"].sum()
    ],
    "DD": [
        df["Calc_DD_Profit"].sum(),
        df["Commission_DD"].sum()
    ]
}, index=["Profit", "Commission"])

commission_vs_profit.T.plot(kind="bar", figsize=(8,5))
plt.title("Commission Drag vs Net Profit")
plt.ylabel("Amount")
plt.show()
```

# Self-Delivery Breakeven Analysis

```python
# delivery per order 

df["SD_CostPerOrder"] = (
    df["SD_DeliveryTotalCost"] / df[sd_orders_col]
)
```

```python
# Breakeven Margin

df["SD_BreakevenFlag"] = np.where(
    df["Margin_SD"] <= 0,
    "Loss Making",
    "Profitable"
)

df["SD_BreakevenFlag"].value_counts()
```

```python
df["SD_BreakevenFlag"].value_counts().plot(
    kind="bar",
    figsize=(6,4),
    title="Self Delivery Profitability Status"
)
plt.show()
```

```python
# Cuisine Profitability Heatmap

cuisine_channel_margin = df.groupby("CuisineType")[[
    "Margin_InStore",
    "Margin_UE",
    "Margin_DD",
    "Margin_SD"
]].mean()

plt.figure(figsize=(10,6))
sns.heatmap(cuisine_channel_margin, annot=True, fmt=".2f", cmap="RdYlGn")
plt.title("Margin Heatmap by Cuisine & Channel")
plt.show()
```

```python
# Top & Bottom Performers
# Worst UE performers
worst_ue = df.nsmallest(10, "Margin_UE")[
    ["RestaurantName", "CuisineType", "Margin_UE"]
]

# Best UE performers
best_ue = df.nlargest(10, "Margin_UE")[
    ["RestaurantName", "CuisineType", "Margin_UE"]
]

worst_ue, best_ue
```
